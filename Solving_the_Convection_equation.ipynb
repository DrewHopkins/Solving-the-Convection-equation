# The convection equation is written as follows

## $\frac{\partial u}{\partial t} + a\frac{\partial u}{\partial x}=0$

 We shall solve this using a numerical scheme consiting of...
 - foward first-order scheme in time
 - backward first-order scheme in space


 From our scratch work, we found that the numerical scheme is as follows...

# $(u_i^{n+1}-u_i^n)/\Delta t + a * (u_i^n-u_{i-1}^n)/\Delta x$ = 0

```python
# Packages
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convectionAnalyticalSolution(x,t,a):
   uu = np.exp(-(x-a*t)**2/0.2**2)
   return uu
```

```python
def getXValues(i):
  #gives the amount of steps in the x interval
  steps_x = int(x.length/i)
  xlower = -2
  rows=[]  
  for w in range(steps_x):
    rows.append(round(xlower + w * i,2))
  return rows
```

```python
def getDeltat(CFL,deltax,a):
  deltat = CFL * deltax / a
  return deltat
```

```python
def numericalScheme(uu,n,i,deltat,xstep,a):
  uuuu = uu[n][i] - a * deltat / xstep * (uu[n][i]-uu[n][i-1])
  return uuuu
```

```python
def intervals():
  x = pd.Interval(left=-2, right=20,closed='both')
  a=1
  tf = 16
  deltax =[0.2,0.1,0.05,0.02]
  CFL = 0.98
  return x,a,tf,deltax,CFL
```

# Plot the results at $t=t_{final}$ for $\Delta x = 0.2$

```python
x,a,tf,deltax,CFL = intervals()
## initial condition
xx = np.linspace(-2,20,300)
u_x_0 = convectionAnalyticalSolution(xx,0,a)
plt.plot(xx,u_x_0, color = 'b', label='intial condition t = 0',linestyle='-.');
## final analytical solution
u_x_f = convectionAnalyticalSolution(xx,tf,a)
plt.plot(xx,u_x_f, color = 'r', label='final analytical solution t = tf',linestyle='-.');
## final numerical solution
xstep = deltax[0]
xxx = getXValues(xstep)
deltat = getDeltat(CFL,xstep,a)
tt = int(tf/deltat)
uuu = []
uu = [uuu]
for b in range(len(xxx)):  
  uuu.append(convectionAnalyticalSolution( xxx[b],0,a )) ## we now have the first vector to the soltion at t = 0
uuu[0]=0
for c in range(tt-1):
  uu.append([0]) ## we now have a vector of vectors or matrix that hold the initial and boundry condition
for n in range(tt-1): 
  for i in range(len(xxx)-1):  
    uu[n+1].append(numericalScheme(uu,n,i+1,deltat,xstep,a))
array_length=len(uu)
plt.plot(xxx,uu[array_length-1],color='g',label='final numerical solution t = tf',linestyle='-.')
plt.title('Convection Wave simmulations $\Delta x = 0.2$')
plt.xlabel('x')
plt.ylabel('u (x,t)')
plt.legend()
plt.show()
```
